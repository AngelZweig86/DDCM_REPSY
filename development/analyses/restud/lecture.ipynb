## State Space

In each period, can choose between four alternatives: (1) Occupation A, (2) Occupation B, (3) School, and (4) Home. There is one additional restriction relevant here. Agents are only able to accumulate ten years of additional education. Afterwards, enrolling in school is simply not an option any more. Even in this restricted model setup, the dimension of the state space increases tremendously over time. In period 40, there are 13,150 states to consider.

## Return to Experience and Schooling

In all of the data sets, occupation two is more skill-intensive in the sense that
schooling has a higher return (except in data set three where they are equal), and "own" experience has a higher return. In addition, experience in occupation one increases skills (and thus the wage) in occupation two while the reverse is never true. Thus, both schooling and occupation one experience provide general skills useful in both occupations, while occupation two experience provides skills specific only to occupation two.

## Choice Patterns



In data set one, the proportion of individuals in occupation one begins at 0.39, rises to a peak in period 5 of 0.46 and gradually declines throughout the rest of the life cycle to 0.23 in period 40.

```python
Image(filename=GRAPH_DIR + '/rslts/data_one/choice_patterns.png')
```

Data set two differs from data set one in several aspects: (1) occupation two has a lower mean wage at $t = 0$ in the second data set than in the first, (2) schooling has positive consumption value, there is a non-zero tuition cost for college and a
larger adjustment cost associated with returning to school in data set two, (3) the value of non-market time is lower in data set two, and (4) the standard deviations of the (ln) wage errors are twice as large, and those of the schooling value and home value errors four times as large, in data set two.

In data set two, the pattern is similar to data set one although the initial rise in participation in occupation one is greater, from 0.34 in period 1 to 0.67 in period 7, and the fall more gradual to 0.56 in period 40.

```python
Image(filename=GRAPH_DIR + '/rslts/data_two/choice_patterns.png')
```

Data set three differs from one and two in that (1) the (ln) wage functions in both occupations are linear in experience, rather than concave, in data set three, (2) the schooling adjustment cost is larger in data set three, (3) the value of home
time is larger in data set three, (4) the standard deviations of the (ln) wage errors are at least twice as big in data set three as in two and at twice as big in data set three as in two and at least four times as big as in one, and the standard
deviations of the schooling and home time value errors are slightly bigger in three than in two, and (5) the ln wage errors of the two occupations have a contemporaneous correlation of $0.5$ and the schooling and home time value errors have a correlation of $-0.5$ in data set three, while all contemporaneous correlations are zero in data sets one and two.

In data set three the rise is even more pronounced, from 0.17 in period 1 to 0.80 in period 12, as the skills acquired in occupation one are more general, and the decline more pronounced, to only 0.27 in period 40, as the returns are obtained by switching to occupation two.

```python
Image(filename=GRAPH_DIR + '/rslts/data_three/choice_patterns.png')
```

Participation in occupation two increases continuously over the life cycle in all three data sets, although more steeply in data sets one and three. The proportion of individuals in school declines rapidly with age in all three data sets as schooling is essentially only an investment good. Schooling has its highest overall return in data set three. Home-time is lowest in data set one and constant over the life cycle, has an inverted u-shape set two reaching a peak of 0.09 in period 7, and has a u-shape in data set three rising to 0.13 in period 40. Thus, data set three generates some degree of voluntary retirement as they have increasingly larger error-variances. The individuals leave occupation one because its investment opportunities are less valuable as the end of the horizon is approached.
