# RESTUD Economy

This notebook visualizes some key economic implications of the simulated economy based on the parameters analyzed by the authors in:

> Michael P. Keane, Kenneth I. Wolpin (1994). The Solution and Estimation of Discrete Choice Dynamic Programming Models by Simulation and Interpolation: Monte Carlo Evidence. Federal Reserve Bank of Minneapolis, No. 181.

> Michael P. Keane, Kenneth I. Wolpin (1994). The Solution and Estimation of Discrete Choice Dynamic Programming Models by Simulation and Interpolation: Monte Carlo Evidence. The Review of Economics and Statistics, 76(4): 648-672.

The original code is written in FORTRAN77, but we upgraded it to FORTRAN90. Both versions are available [here](https://github.com/robustToolbox/package/tree/master/development/analyses/restud/codes). The solution and simulation of the economies, as well as the creation of all graphs below, is straightforward.
