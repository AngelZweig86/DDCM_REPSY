```python
# standard library
import shlex
import glob
import sys
import os

# PYTHONPATH
sys.path.insert(0, os.environ['ROBUPY'])

# project package
from robupy import read


robupy_obj = read('../model.robupy.ini')

num_periods = robupy_obj.get_attr('num_periods')


def get_ambiguity_levels():
    """ Infer ambiguity levels from directory structure.
    """
    
    levels = []

    for level in glob.glob('../*/'):
        
        # Cleanup strings
        level = level.replace('../', '').replace('/', '')
        
        # Skip directory
        if level in ['bin', 'graphs']:
            continue

        # Collect levels
        levels += [level] 

    # Finishing
    return sorted(levels)
        

# Get all ambiguity levels
levels = get_ambiguity_levels()

# Read choice distribution in final state by ambiguity level
labels = ['Occupation A', 'Occupation B', 'Schooling', 'Home']

# Initialize dictionary
shares = dict()
for label in labels:
    shares[label] = []

for level in levels:
    with open('../' + level + '/data.robupy.info', 'r') as output_file:
        for line in output_file.readlines():
            # Split lines
            list_ = shlex.split(line)
            # Skip empty lines
            if not list_:
                continue
            # Extract shares    
            if (str(num_periods) in list_[0]):
                for i, label in enumerate(labels):
                    shares[label] += [float(list_[i + 1])]

print()
```

```python
# plotting
%matplotlib inline

import matplotlib.pylab as plt
import numpy as np

from matplotlib.ticker import FuncFormatter
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Initialize plot
ax = plt.figure(figsize=(12,8)).add_subplot(111)

styles  = ['--k', '-k', '*k', ':k']
# Draw lines
for i, key_ in enumerate(shares.keys()):
    ax.plot(levels, shares[key_], styles[i], label=key_)
    
# Both axes
ax.tick_params(axis='both', right='off', top='off')

# x-axis
ax.set_xticklabels(ax.get_xticks().astype(int), fontsize=20)
ax.set_xlabel('Ambiguity', fontsize=20)

# y-axis
ax.set_ylabel('Shares', fontsize=20)

# legend
#handles, labels = ax.get_legend_handles_labels()
#plt.legend(loc='upper_center', handles, labels, ncol=2, frameon=False, fontsize=16)

# Set up legend
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.10),
    fancybox=False, frameon=False, shadow=False, ncol=4,
    fontsize=20)

# Write out to
plt.savefig('choices_ambiguity.pdf', bbox_inches='tight', format="pdf")
```
