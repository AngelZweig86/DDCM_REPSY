This notebook visualizes some key economic implications of the simulated economy based on the first set of parameters analyzed by the authors.

**Parametrization**

```python
coeffs = {}
coeffs['A'] = [9.21, 0.038, 0.033, -0.0005, 0.000, -0.0000]
coeffs['B'] = [8.48, 0.070, 0.022, -0.0005, 0.067, -0.0010] 
```

**Functions**

```python
def wage_function(edu, exp_A, exp_B, coeffs):
    """ This function calculates the wage based on agent's
    covariates for a given parameterization.
    """
    
    # Intercept
    wage = coeffs[0]

    # Schooling
    wage += coeffs[1] * edu

    # Experience A
    wage += coeffs[2] * exp_A
    wage += coeffs[3] * exp_A ** 2

    # Experience B
    wage += coeffs[4] * exp_B
    wage += coeffs[5] * exp_B ** 2

    # Transformation
    wage = np.exp(wage)
    
    # Finishing
    return wage
```

```python
def return_to_education(edu, coeffs, which):
    """ This function evaluates.
    
        TODO: Set experience to AVEWRAGE level
        
    """
    assert (which in ['A', 'B'])
    

    # Distribute covariates
    exp_A, exp_B = 10, 10 
    
    # Get wage
    wage = wage_function(edu, exp_A, exp_B, coeffs[which])


    # Finishing
    return wage
```

```python
xvals = range(10, 20)

yvals = dict()

for which in ['A', 'B']:
    yvals[which] = []
    for edu in xvals:
        yvals[which] += [return_to_education(edu, coeffs, which)]

ax = plt.subplot()   
ax.plot(xvals, yvals['A'], '-k', label='Occupation A')
ax.plot(xvals, yvals['B'], '--k', label='Occupation B')
     
# Both axes
ax.tick_params(axis='both', right='off', top='off')
     
# x-axis
ax.set_xticklabels(ax.get_xticks().astype(int), fontsize=18)
ax.set_xlabel('Years of Education', fontsize=18)
 
# y-axis
yticks = ['{:,.0f}'.format(y) for y in ax.get_yticks().astype(int)]
ax.set_yticklabels(yticks, fontsize=18)
ax.set_ylabel('Rewards', fontsize=18)
 
# legend
handles, llabels = ax.get_legend_handles_labels()
legend = plt.legend(handles, llabels, ncol=2, frameon=False, \
                        fontsize=16)






plt.show()
# legend.set_bbox_to_anchor([-.15, -.15, 1.0, 0.03])
#plt.savefig('{}.{}'.format(fname, fmat), bbox_inches='tight', loc=1, \
#                format=fmat, bbox_extra_artists=(legend,))
#plt.clf()
```
